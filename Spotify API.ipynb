```python
import pandas as pd
import spotipy
from spotipy.oauth2 import SpotifyClientCredentials
```

```python
sp = spotipy.Spotify(client_credentials_manager=SpotifyClientCredentials())
```

```python
song_data = pd.read_csv("songdata.csv")
song_data
```

```python
def get_explicit_label(artist, song):
    label = ''
    results = sp.search(q=song, type='track', limit=5)
    if len(results['tracks']['items'])>0:
        #print(1)
        for t in results['tracks']['items']:
            if t['artists'][0]['name'] == artist:
                #print(2)
                label = t['explicit']
                break
            else:
                #print(3)
                label = 'no match'
    else:
        #print(4)
        label = 'no match'
    print(label)
    return label
```

```python
label = []
for i in range(0,len(song_data)):
    label.append(get_explicit_label(song_data['artist'][i], song_data['song'][i]))
```

```python
explicit_label = pd.DataFrame(label, columns=['explicit_label']) 
```

```python
sub_song_data = pd.concat([song_data, explicit_label], axis=1)
sub_song_data
```

```python
sub_song_data.to_csv("subsongdata_57650.csv", header=True)
```
